# Problem Statement
When processing Excel files, we would like to see a quick snapshot of differences between any 2 Excel files.

# Solution
This work book exports a difference Excel file indicating the differences between the two Excel inputs.

# Code
### 1. Specify inputs

```python
import pandas as pd
from pathlib import Path
```

```python
first_excel = Path('/Volumes/RadReportsAdmin/CR Folder/Critical Results - March 2025.xlsx')
second_excel = Path('/Users/maiyn/Work/Packages/RAD_funcs/Processed - Critical Results - March 2025.xlsx')

df1 = pd.read_excel(first_excel)
df2 = pd.read_excel(second_excel)
```

```python
# Get diff
common_column = 'Sort#'
merged = df1.merge(df2, how='outer', on='Sort#', suffixes=('_original', '_processed'))
```

```python
# Get relevant diff columns
relevant_column = 'CR called Date/Time'
diff = merged[merged[f'{relevant_column}_original'] != merged[f'{relevant_column}_processed']]
diff
```

```python
# Define columns to output
columns = [
'Sort#', 
'Report Created Date/Time_original', 
'CR called Date/Time_original',	
'CR called Date/Time_processed', 
"Minutes From\nCreated to Called_original", 
"Minutes From\nCreated to Called_processed", 
"Note",
"Report Text_processed"
]
diff[columns].to_excel('Diff.xlsx', index=False)
```
